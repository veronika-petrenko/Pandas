```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("bank.csv", delimiter=";")
df
```

### 1. Eliminar las columnas: default, balance, housing, loan, duration, campaign, pdays, previous, poutcome

```python
df.drop('default','balance','housing','loan','duration','campaign','pdays','previous','poutcome',axis=1)
```

```python
df = df.drop(df.loc[:,'default':'loan'],axis=1)
```

```python
df = df.drop(df.loc[:,'duration':'poutcome'],axis=1)
```

```python
df
```

### 2. Aplicar map sobre la columna contact y cambiar cada elemento a español

```python
df['contact'].unique()
```

```python
df['contact'] =df['contact'].map({'cellular':'movil','unknown':'desconocido','telephone':'telefono'})
```

```python
df
```

### 3. Aplicar applymap sobre las columnas job y marital y pasa cada elemento a mayusculas

```python
df[['job','marital']] = df[['job','marital']].applymap(lambda i:i.upper())
df
```

```python
df10 = df.groupby(['job'])['age'].min()
df10
```
